# Data Visualization

```python
labels = ['Malignant', 'Benign']
plt.figure(figsize=(10, 6))
sns.countplot(x='diagnosis', data= breast_cancer, hue='diagnosis')
plt.xticks([0, 1],labels)
plt.legend(labels, title='Diagnosis')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=breast_cancer, x='radius_mean', hue='diagnosis', kde=True)
plt.legend(labels, title='Diagnosis')
plt.show()
```

```python
top_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']
plt.figure(figsize=(12, 8))
for i, feature in enumerate(top_features, 1):
  plt.subplot(2, 2, i)
  sns.boxplot(x = 'diagnosis', y=feature, data= breast_cancer, palette='Set2', hue='diagnosis', legend=False) # Added hue='diagnosis' and legend=False
  plt.title(f'{feature} by diagnosis')

plt.tight_layout()
plt.show()
```

```python
top_features = ['radius_mean', 'perimeter_mean', 'area_mean', 'diagnosis']
plt.figure(figsize=(12, 8))
sns.pairplot(breast_cancer[top_features],  hue='diagnosis', diag_kind='kde')
plt.suptitle('Pairplot of Top Features by Diagnosis', y=1.02)

plt.tight_layout()
plt.show()
```

```python
labels = ['Malignant', 'Benign']
plt.figure(figsize=(10, 6))

counts = breast_cancer['diagnosis'].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%',wedgeprops=dict(width=0.5), startangle= -40)
plt.title('Diagnosis Class Proportions')
plt.legend(labels, title='Diagnosis', loc='center')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.violinplot(x='diagnosis', y='texture_mean', data=breast_cancer, hue='diagnosis')
plt.title('Texture Mean by Diagnosis')
plt.show()
```

```python
# Drop non-numeric columns like 'diagnosis
numeric_data = breast_cancer.select_dtypes(include=[np.number])
numeric_data.drop('id', axis=1,  inplace=True)
plt.figure(figsize=(20, 18))
sns.heatmap(numeric_data.corr(),
            annot=True,
            linewidths=0.5,
            cmap='Purples',
            square=True,
            cbar_kws={"shrink": 0.5},
            fmt=".2f")
plt.show()
```
